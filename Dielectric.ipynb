# Cálculo de la densidad superficial de carga y del campo eléctrico
Usemos SymPy para calcular el gradiente del potencial electrostático para una carga puntual unitaria con carga q frente a un medio con permitividad eléctrica epsilon. Comenzamos definiendo símbolos para el potencial, las coordenadas espaciales (x,y,z), y otras magnitudes de interés.

```python
# Definimos variables simbólicas reales phi (para el potencial), x,y,z para las coordenadas, y x',y',z' para las coordenadas primadas
phi,q,d  = symbols("phi q d", real=True)
x,y,z    = symbols("x y z",   real=True)
epsilon  = symbols("epsilon",   real=True)
```

```python
# Definimos el potencial y lo imprimimos
phi = Piecewise(
    (2/(1+epsilon) * q/(sqrt(x**2+y**2+(z-d)**2)), z<0),
    (q/(sqrt(x**2+y**2+(z-d)**2)) + (1-epsilon)/(1+epsilon) * q/(sqrt(x**2+y**2+(z+d)**2)), z>=0))
phi
```

```python
# La densidad superficial de carga es (Ez(0+)-Ez(0-))/(4pi), donde Ez = -d phi/dz
sigma = -(1/(4*pi)) * (diff(q/(sqrt(x**2+y**2+(z-d)**2)) + (1-epsilon)/(1+epsilon) * q/(sqrt(x**2+y**2+(z+d)**2)),z) -
                       diff(2/(1+epsilon) * q/(sqrt(x**2+y**2+(z-d)**2)),z)).subs(z,0)
simplify(sigma)
```

```python
# Calculemos las componentes del campo eléctrico. Comenzamos por E_x = -d phi/dx:
E_x = -diff(phi, x)
E_x
```

```python
# E_y = -d phi/dy:
E_y = -diff(phi, y)
E_y
```

```python
# E_z = -d phi/dz:
E_z = -diff(phi, z)
E_z
```

# Potencial electrostático y campo eléctrico en el plano y=0
Ahora usamos NumPy y Matplotlib para graficar estos resultados. Usamos la función "lambdify" de SymPy para convertir las expresiones simbólicas en funciones numéricas para NumPy, y generar así expresiones para las cantidades de interés que queremos graficar reemplazando antes y=0, y dejando a la carga q, a la distancia d al plano, y a la permitividad eléctrica epsilon como variables.

```python
sigma     = lambdify([x,y,q,d,epsilon],sigma)
potential = lambdify([z,x,q,d,epsilon],phi.subs(y,0))
ez        = lambdify([z,x,q,d,epsilon],E_z.subs(y,0))
ex        = lambdify([z,x,q,d,epsilon],E_x.subs(y,0))
```

```python
# Importamos NumPy y Matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos valores para la carga y la distancia de la carga al plano dieléctrico
char = 1. # carga (en unidades adimensionales)
dist = 5. # distancia de la carga al plano (prueben cambiar estos valores)
epsi = 6. # permitividad eléctrica del dieléctrico
```

```python
# Graficamos la densidad superficial de carga (de polarización)
x = np.arange(-10,10,.5)
y = np.arange(-10,10,.5)
xmesh,ymesh = np.meshgrid(x,y)

fig = plt.figure(figsize=(7,7))
ax  = fig.gca(projection='3d')
ax.plot_surface(xmesh,ymesh,sigma(xmesh,ymesh,char,dist,epsi),cmap='viridis')
plt.xlabel(r'$x$',fontsize=16)
plt.ylabel(r'$y$',fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Graficamos el valor del potencial y las curvas equipotenciales en el plano (x,z)
z = np.arange(-10.05,20.05,.1) # Desplazamos la malla en 0.05 para evitar la singularidad en la posición de la carga
x = np.arange(-10.05,10.05,.1)
zmesh,xmesh = np.meshgrid(z,x)

plt.figure(figsize=(9,6))
plt.pcolormesh(zmesh,xmesh,potential(zmesh,xmesh,char,dist,epsi),vmin=0,vmax=char,shading='auto')
plt.contour(z,x,-np.log(potential(zmesh,xmesh,char,dist,epsi)),colors=['white'],levels=30)
plt.plot([0.,0.],[np.min(x),np.max(x)],'r')
plt.xlabel(r'$z$',fontsize=16)
plt.ylabel(r'$x$',fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Graficamos el campo eléctrico
z = np.arange(-10,20,1)
x = np.arange(-10,10,1)
zmesh,xmesh = np.meshgrid(z,x)

plt.figure(figsize=(9,6))
plt.streamplot(zmesh,xmesh,ez(zmesh,xmesh,char,dist,epsi),ex(zmesh,xmesh,char,dist,epsi),)
plt.xlabel(r'$z$',fontsize=16)
plt.ylabel(r'$x$',fontsize=16)
plt.plot([0.,0.],[np.min(x),np.max(x)],'r')
plt.tight_layout()
plt.show()
```

```python
# Veamos finalmente lo que ocurre en el salto en la interfaz
z = np.arange(-10.015,20.015,.01)
x = 2.

eps = np.heaviside(z,0) + epsi*np.heaviside(-z,1)

plt.figure(figsize=(9,6))
plt.plot(z,ez(z,x,char,dist,epsi),'r',label=r'$E_z$')
plt.plot(z,eps*ez(z,x,char,dist,epsi),'r--',label=r'$D_z$')
plt.plot(z,ex(z,x,char,dist,epsi),'b',label=r'$E_x$')
plt.plot(z,eps*ex(z,x,char,dist,epsi),'b--',label=r'$D_x$')
plt.legend(fontsize=16)
plt.xlabel(r'$z$',fontsize=16)
plt.ylabel(r'$x$',fontsize=16)
plt.tight_layout()
plt.show()
```
